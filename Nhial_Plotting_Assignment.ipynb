```python
plt.title("Distribution of carbon dioxide")
plt.hist(fuel_econ_df['co2'], bins=20, color="skyblue", log=True, edgecolor="black")
plt.xlabel("CO2 Emissions (g/mile)")
plt.ylabel("frequency")
plt.grid()
plt.show()
```

```python
# Check data types of all columns
print(fuel_econ_df.info())

# Select only numeric columns
numeric_data = fuel_econ_df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
plt.figure(figsize=(10, 8))
correlation_matrix = numeric_data.corr()  # Only numeric columns are used
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
